```python
import pandas as pd 
import numpy as np 


temp_data = pd.read_csv('temp_data.csv', index_col=0)
temp_data 
```

```python
def value_function(data, temp_devices, humidity_devices, light_devices):
    
    cols = [f'temp_{int(d)}' for d in temp_devices]
    cols += [f'humidity_{int(d)}' for d in humidity_devices]
    cols += [f'light_{int(d)}' for d in light_devices]
    
    data_slice = data[cols]
    
    
    total_size = data_slice.shape[0]
    print(f'Total size {total_size}')
    
    
    
    freqs = np.array(list(dict(data_slice.groupby(cols).size()).values()))
    assert np.sum(freqs) == total_size
    freqs = np.array(freqs)
    freqs = freqs[np.nonzero(freqs)]
    freqs = freqs / total_size
    
    entropy = - np.sum(freqs * np.log(freqs))
    
    
    return entropy



select_devices = []
for col in temp_data.columns:
    
    select_devices.append(len(temp_data[col].dropna()))

select_devices = np.array(select_devices)
device_indices = np.argsort(-select_devices)[:20]

select_cols = [temp_data.columns[idx] for idx in device_indices] 

select_temp_data = temp_data[select_cols].dropna() 
select_temp_data
```

```python
min_temp, max_temp = select_temp_data.min().min(), select_temp_data.max().max()
bin_width = 2
labels = list(range(int(min_temp), int(max_temp)+1, bin_width))

print(len(labels))

# data['temperature_binned'] = 
for c in select_temp_data.columns:
    select_temp_data[c] = pd.cut(select_temp_data[c], bins=len(labels), labels=labels)
    


# freqs = np.array(list(dict(select_temp_data.groupby(cols).size()).values()))
# freqs 
```

```python
vals = select_temp_data.groupby(['temp_47','temp_48','temp_45','temp_21', 'temp_21','temp_23', 'temp_46', 'temp_36']).size().values
vals = vals[vals.nonzero()]
vals = vals / np.sum(vals)

H = -np.sum(vals * np.log(vals))
print(H)
```
